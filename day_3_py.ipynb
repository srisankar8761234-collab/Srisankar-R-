```python
for i in range(1,6):
  for j in range(5):
    print(i,end="")
  print()
```

```python
for i in range(5,0,-1):
  for j in range(i):
    print("*",end="")
  print()
```

```python
# hollow square
for i in range(1,6):
  for j in range(1,6):
    if(i==1 or i==5 or j==1 or j==4):
      print("*",end="")
    elif(i==2 and j==1)or(i==3 and j==1)or(i==4 and j==1):
      print(".",end="")
    else:
      print(" ",end="")
  print()
```

```python
# x pattern
for i in range(5):# 0 0=0
  for j in range(5):#0 1=1
    if( j==5-i-1 or i==j ):# 1 0=1
      print("*",end="")#11=1
    else:
      print(" ",end="")
  print()
```
