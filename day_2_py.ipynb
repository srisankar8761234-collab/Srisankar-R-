```python
# employee id card
def employee(n):
  for i in range(n):
    name=input("enter your name: ")
    age=int(input("enter your age: " ))
    dept=input("enter your department: ")
    salary=float(input("enter your salary: "))
    print("name:",name)
    print("age:",age)
    print("department:",dept)
    print("salary:",salary)
x=int(input("enter no of employee:"))
print("employee id card")
employee(x)
```

```python
#grocery bill calculator
def bill():
  pen=int(input("enter cost of item:"))
  scale=int(input("enter cost of item"))
  note=int(input("enter cost of item"))
  pencil=int(input("enter cost of item"))
  marker=int(input("enter cost of item"))
  total=pen+scale+note+pencil+marker
  print("the total amount:",total)
bill()
```
